```python
import pandas as pd
```

```python
df = pd.read_csv('../data/raw/covid.csv', sep=',')
```

# Validate NULLS

```python
validate_nulls = df.isna().sum().to_frame()
validate_nulls.reset_index(inplace=True)
validate_nulls.columns = ['columns','null_total']
validate_nulls['total_rows'] = df.shape[0]
validate_nulls['percentage_nulls'] =  round(validate_nulls['null_total'] / validate_nulls['total_rows'], 5) * 100
validate_nulls
```

```python
nulls_city = df[df['city'].isna()]
nulls_city.head(2)
```
